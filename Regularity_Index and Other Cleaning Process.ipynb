# Regularity Index Calculation

```python
# Calculate the total number of commits for each product and filter out products with less than 1000 commits
product_commit_counts = df5['product'].value_counts()
products_to_keep = product_commit_counts[product_commit_counts >= 1000].index
df5 = df5[df5['product'].isin(products_to_keep)]
```

```python
df5['commit_time'] = pd.to_datetime(df5['commit_time'], errors='coerce')
df5 = df5.dropna(subset=['commit_time'])


# Calculate the total number of days for each repository
repo_durations = df5.groupby('product')['commit_time'].agg(lambda x: (x.max() - x.min()).days + 1).reset_index(name='total_days')

# Calculate commits per day for each developer
df5['commit_date'] = df5['commit_time'].dt.date
commits_per_day = df5.groupby(['product', 'committer', 'commit_date']).size().reset_index(name='commits')

# Define the function to calculate the regularity index
def calculate_regularity_index(committer_data, total_days):
    di = committer_data['commit_date'].nunique()
    C = committer_data['commits'].sum()
    avg_commits_per_day = C / di
    commit_sd = np.sqrt((1 / di) * np.sum((committer_data['commits'] - avg_commits_per_day) ** 2))
    R = (di / total_days) * (1 / (commit_sd + 1))
    return pd.Series([R, di, C], index=['regularity_index', 'active_days', 'total_commits'])

# Create an empty DataFrame to store the results
all_results = pd.DataFrame()

# Create an empty DataFrame to store results for products without all three categories
incomplete_categories = pd.DataFrame()

# Process each product separately
for product in df5['product'].unique():
    product_data = df5[df5['product'] == product]
    
    # Calculate commits per day for the current product
    product_commits_per_day = product_data.groupby(['committer', 'commit_date']).size().reset_index(name='commits')
    
    # Calculate regularity index for each committer in the current product
    product_regularity_indices = product_commits_per_day.groupby('committer').apply(
        lambda x: calculate_regularity_index(x, repo_durations.loc[repo_durations['product'] == product, 'total_days'].values[0])
    ).reset_index()
    
    # Sort by regularity index in descending order
    product_regularity_indices = product_regularity_indices.sort_values(by='regularity_index', ascending=False)
    
    # Initialize 'category' column
    product_regularity_indices['category'] = 'medium'
    
    if len(product_regularity_indices) >= 3:
        # Calculate the 30th and 75th percentiles
        P30 = np.percentile(product_regularity_indices['regularity_index'], 30)
        P75 = np.percentile(product_regularity_indices['regularity_index'], 75)
        
        # Classify committers into high, low, and medium based on percentiles
        product_regularity_indices.loc[product_regularity_indices['regularity_index'] < P30, 'category'] = 'low'
        product_regularity_indices.loc[product_regularity_indices['regularity_index'] > P75, 'category'] = 'high'
        
        # Ensure that all three categories are present
        counts = product_regularity_indices['category'].value_counts()
        if 'low' not in counts or 'high' not in counts or 'medium' not in counts:
            min_r = product_regularity_indices['regularity_index'].min()
            max_r = product_regularity_indices['regularity_index'].max()
            product_regularity_indices.loc[product_regularity_indices['regularity_index'] == min_r, 'category'] = 'low'
            product_regularity_indices.loc[product_regularity_indices['regularity_index'] == max_r, 'category'] = 'high'
            remaining_indices = product_regularity_indices[
                (product_regularity_indices['category'] != 'low') & 
                (product_regularity_indices['category'] != 'high')
            ].index
            product_regularity_indices.loc[remaining_indices, 'category'] = 'medium'
    
    else:
        # If less than 3 committers, assign categories manually
        if len(product_regularity_indices) == 1:
            product_regularity_indices.iloc[0, product_regularity_indices.columns.get_loc('category')] = 'high'
        elif len(product_regularity_indices) == 2:
            product_regularity_indices.iloc[0, product_regularity_indices.columns.get_loc('category')] = 'high'
            product_regularity_indices.iloc[1, product_regularity_indices.columns.get_loc('category')] = 'low'
    
    # Add the product column
    product_regularity_indices['product'] = product
    
    # Append the result to the final DataFrame
    all_results = pd.concat([all_results, product_regularity_indices], ignore_index=True)
    
    # Check if the product has all three categories
    if 'low' not in product_regularity_indices['category'].values or \
       'medium' not in product_regularity_indices['category'].values or \
       'high' not in product_regularity_indices['category'].values:
        incomplete_categories = pd.concat([incomplete_categories, product_regularity_indices], ignore_index=True)

# Save the regularity indices for products without all three categories to a CSV file
incomplete_categories.to_csv('incomplete_categories.csv', index=False)

# Save the regularity indices for all developers in all products to a CSV file
all_results.to_csv('all_results.csv', index=False)
```
